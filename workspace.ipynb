```python
from pathlib import Path
import shutil

model_names = Path("./model.txt").read_text().splitlines()

for dir in Path("./0510/PF/").iterdir():
    for sub_dir in dir.iterdir():
        if sub_dir.stem not in model_names:
            print(sub_dir.stem, "not in list!")
            shutil.rmtree(sub_dir)
            continue
        file_names = [file.stem for file in sub_dir.iterdir()]
        if sub_dir.stem in file_names:
            continue
        else:
            if f"{sub_dir.stem}-1" in file_names:
                path = sub_dir / f"{sub_dir.stem}-1.jpg"
            elif f"{sub_dir.stem}-A" in file_names:
                path = sub_dir / f"{sub_dir.stem}-A.jpg"
            else:
                print(sub_dir.stem)
            assert path.is_file()
            extension = path.suffix
            directory = path.parent
            new_name = sub_dir.stem + extension
            path.rename(Path(directory, new_name))
```

```python
model_names = Path("./model.txt").read_text().splitlines()

for dir in Path("./0510/PF/").iterdir():
    for sub_dir in dir.iterdir():
        for file in sub_dir.iterdir():
            if file.stem == sub_dir.stem:
                continue
            end = file.stem.split("-")[-1]
            assert file.stem.endswith(f"-{end}"), file.stem
            extension = file.suffix
            directory = file.parent
            new_name = f"{sub_dir.stem}({end}){extension}"
            file.rename(Path(directory, new_name))
```
